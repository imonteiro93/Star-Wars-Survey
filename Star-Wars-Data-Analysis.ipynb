## Cleaning the data set

```python
#removing all rows, where RespondentID is not a number
star_wars = star_wars[pd.notnull(star_wars['RespondentID'])]
star_wars
```

For the second and third columns, which are of Yes/No answer (or NaN), they may be turned into booleans for easier analysis later. Booleans are easier to use since they don't need string comparison to select rows and columns.

```python
#mapping dictionary

yes_no = {
    "Yes": True,
    "No": False
}

star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)
#star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)]
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)
```

```python
movies_map = {
    "Star Wars: Episode I  The Phantom Menace": True,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True,
    np.NaN: False
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movies_map)
```

```python
new_cols = {'Which of the following Star Wars films have you seen? Please select all that apply.':'saw_ep_1', 'Unnamed: 4':'saw_ep_2','Unnamed: 5':'saw_ep_3','Unnamed: 6':'saw_ep_4','Unnamed: 7':'saw_ep_5','Unnamed: 8':'saw_ep_6'}

star_wars = star_wars.rename(columns=new_cols)
star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float) #making movie ranking values as floats
rank_map = {'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.':'ep_1_rank','Unnamed: 10':'ep_2_rank','Unnamed: 11':'ep_3_rank','Unnamed: 12':'ep_4_rank','Unnamed: 13':'ep_5_rank','Unnamed: 14':'ep_6_rank'}
star_wars = star_wars.rename(columns=rank_map)
star_wars.head()
```

## Segmented analysis
### Gender differences

There are different columns that allow us to split the analysis in groups by creating subsets of that column. We can for example split the 'Gender' column into **Male** and **Female** and than compare the highest rated or most viewed movies.

```python
males = star_wars[star_wars['Gender'] == 'Male']
females = star_wars[star_wars['Gender'] == 'Female']
```

```python
import matplotlib.pyplot as plt

males_mean = males[males.columns[9:15]].mean()
females_mean = females[females.columns[9:15]].mean()

fig, ax = plt.subplots()
male_plot = ax.bar(np.arange(6), males_mean,width=0.35, color='r')
female_plot = ax.bar(np.arange(6)+0.35, females_mean, width=0.35, color='b')

#text for labels, title and axes ticks
ax.set_ylabel('Mean Rank')
ax.set_title('Ranking by Gender')
ax.set_xticks(np.arange(6) + 0.35 / 2)
ax.set_xticklabels(('Ep_I', 'Ep_II', 'Ep_III', 'Ep_IV', 'Ep_V','Ep_VI'))
plt.show()
```

```python
males_sum = males[males.columns[9:15]].sum()
females_sum = females[females.columns[9:15]].sum()

fig, ax = plt.subplots()
male_plot = ax.bar(np.arange(6), males_sum,width=0.35, color='r')
female_plot = ax.bar(np.arange(6)+0.35, females_sum, width=0.35, color='b')

#text for labels, title and axes ticks
ax.set_ylabel('Views')
ax.set_title('Views by Gender')
ax.set_xticks(np.arange(6) + 0.35 / 2)
ax.set_xticklabels(('Ep_I', 'Ep_II', 'Ep_III', 'Ep_IV', 'Ep_V','Ep_VI'))
plt.show()
```

From the plots above, we can see that much more men watched the second trilogy (Episodes I-III) than women. This may show that men had more interest to watch the movies. Interestingly, men ranked these newer movies lower than women did.
